## Melakukan Data Cleaning

```python
from sklearn.impute import SimpleImputer
```

```python
df.isna().sum()
```

## Merubah Nilai Null Yang Punya Tipe Data Desimal

```python
imputer = SimpleImputer(strategy = 'mean')
df["energy"] = imputer.fit_transform(df[["popularity"]])
print("\nJumlah null : ")
print(df.isna().sum())
```

## Merubah Nilai Null Yang Punya Tipe Data Bilangan Bulat

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["popularity"] = imputer.fit_transform(df[["popularity"]])
print("\nJumlah null : ")
print(df.isna().sum())
```

## Merubah Nilai Null Yang Punya Tipe Data Kategorikal

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["explicit"] = imputer.fit_transform(df[["explicit"]])
print("\nJumlah null : ")
print(df.isna().sum())
```

## Melakukan Data Cleaning Pada Data Dengan Nilai Duplikat

## Melakukan One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
encoder = OneHotEncoder(sparse=False) 
explicitEnc = encoder.fit_transform(df[["explicit"]]) 
explicitEnc = pd.DataFrame(explicitEnc) 
df = df.join(explicitEnc) 
df 
```
