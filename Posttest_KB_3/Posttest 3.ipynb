# Grafik Pie

```python
df["gender"].value_counts()
```

```python
data_pie = df["gender"].value_counts().rename_axis("gender").reset_index(name="id_count")
```

```python
data_pie.head()
```

```python
plt.title("perbandingan gender")
plt.pie(data_pie["id_count"],labels=data_pie["gender"],startangle=90,autopct='%.1f%%')
plt.show()
```

#### dari hasil grafik perbandingan gender yaitu M = 52,5% dan W = 47,5%
